```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o-')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("weight")
plt.ylabel("Age")
plt.title("My Graph")
plt.legend(['Age'], loc=1)
```

# Multiple plots in a single Canvas

```python
plt.subplot(121)#row,col,plotnumber
plt.plot(x,y,'r')

plt.subplot(122)
plt.plot(y,x,'b')
```

```python
plt.subplot(221)#row,col,plotnumber
plt.plot(x,y,'r')

plt.subplot(222)
plt.plot(y,x,'b')

plt.subplot(223)#row,col,plotnumber
plt.plot(x,y,'y')

plt.subplot(224)
plt.plot(y,x,'g')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
for row in axes:
    for col in row:
        col.plot(x,y)
plt.tight_layout()
```

```python
x = range(10)
y = range(10)

fig, ax = plt.subplots(nrows=2, ncols=2)

for row in ax:
    for col in row:
        col.plot(x, y)

plt.show()
```
